# 【問題3】標準化クラスをスクラッチで作成

- 理解をより深めるため、StandardScalerをスクラッチで作成する。

- scikit-learnは使わず、NumPyなどを活用して標準化の計算を記述する。（具体的にはfitメソッドとtransformメソッドを作る）


クラスの作成方法は関数に近く、メソッドはクラスの中にさらにインデントを一段下げて記述する。

- インスタンス変数を作成する際はself.mean_のようにselfを付ける。
- クラスの外からscaler.mean_と書いていたscalerの部分が自分自身を表すselfになっている。

```python
class ScratchStandardScaler():
    """
    標準化のためのクラス
    
    Attributes
    -----
    mean_: 次の形のndarray, shape(n_features,)
        平均
    var_: 次の形のndarray, shape(n_features,)
        分散
    """
    
    def fit(self, X):
        """
        標準化のために平均と標準偏差を計算する。
        
        Parameters
        ----------
        X : 次の形のndarray, shape(n_samples, n_features)
            学習データ
        
        """
        
        self.mean_ = sum(X) / len(X) # 平均 = 配列の値の合計 / 配列の値の個数
        self.var_ = sum([(x - self.mean_)**2 for x in X]) / len(X) # 分散 ＝ (配列の各値と平均の差の二乗) / 配列の値の個数
        
        pass
    
    def transform(self, X):
        """
        fitで求めた値を使い標準化を行う。
        
        Parameters
        ----------
        X : 次の形のndarray, shape(n_samples, n_features)
            特徴量
        
        Returns
        ----------
        X_scaled : 次の形のndarray, shape(n_samples, n_features)
            標準化された特徴量
        
        """
        
        X_scaled = (X - self.mean_) / (self.var_** (1/2)) # 標準化： （配列の値 - 平均） / 標準偏差（＝分散の平方根）

        pass
        
        return X_scaled
    
```

```python
# 実際に定義したクラスを用いて標準化してみる
import numpy as np
from sklearn.datasets import load_iris

data = load_iris()
X = data.data[:10]

scratch_scaler = ScratchStandardScaler()
scratch_scaler.fit(X)
print("平均 :", scratch_scaler.mean_)
print("分散 :", scratch_scaler.var_)
X_std = scratch_scaler.transform(X)
```

# 【課題4】 四則演算を行うクラスの作成

- ExampleClassは足し算のメソッドを持っているが、これに引き算、掛け算、割り算のメソッドを加える。

- また、コンストラクタに入力されたvalueが文字列や配列など数値以外だった場合には警告文を出し、self.value=0とするコードを追加する。

- クラス名や説明文も適切に書き換える。

```python
# 初期クラス
class ExampleClass():
    """
    説明用の簡単なクラス
    
    Parameters
    ----------
    value: float or int
        初期値
        
    Attributes
    ----------
    value: float or int
        計算結果
    """
    
    def __init__(self, value):
        self.value = value
        print("初期値{}が設定されました".format(self.value))
        
    def add(self, value2):
        """
        受け取った引数をself.valueに加える
        """
        self.value += value2
```

```python
example = ExampleClass(5)
print("value:", example.value)
```

```python
example.add(3)
print("value:", example.value)
```

```python
# 初期クラスに引き算、掛け算、割り算のメソッドを追加する
# コンストラクタに入力されたvalueが文字列や配列など数値以外だった場合に警告文を出し、self.value = 0とする。
class ExampleClass():
    """
    説明用の簡単なクラス
    
    Parameters
    ----------
    value: float or int
        初期値
        
    Attributes
    ----------
    value: float or int
        計算結果
    """
    
    def __init__(self, value=True):
        """
        受け取った引数が数値以外の場合、警告文を出す
        """
        if (type(value) == int) or (type(value) == float):
            self.value = value
            print("初期値{}が設定されました".format(self.value))
        else:
            print("引数には数値を入力してください")
            self.value = 0
            print("便宜的に初期値{}が設定されました".format(self.value))
        
    def add(self, value2):
        """
        受け取った引数をself.valueに加える
        """
        self.value += value2
        
    def subtract(self, value3):
        """
        受け取った引数をself.valueから引く
        """
        self.value -= value3
    
    def multiply(self, value4):
        """
        受け取った引数をself.valueに乗じる
        """
        self.value *= value4
    
    def divide(self, value5):
        """
        受け取った引数でself.valueを割る
        """
        self.value /= value5
    
```

```python
# 警告文の確認
sample_str = "sample"
sample_list = [1,2]

example1 = ExampleClass(sample_str) # ExampleClassに文字列を渡す
print("")
example2 = ExampleClass(sample_list) # ExampleClassにリストを渡す
```

```python
# 四則演算の確認

example3= ExampleClass(10)
print("")

# 
subtract_factor = 5
multiply_factor = 4
divide_factor = 10

# メソッドが機能することを確認する
example3.subtract(subtract_factor) # 　２を引く（結果は5）
print("{0}を引いた結果は：{1}".format(subtract_factor, example3.value))
print("")

example3.multiply(multiply_factor) # 4を乗じる（結果は20）
print("さらに{0}を乗じた結果は：{1}".format(multiply_factor, example3.value))
print("")

example3.divide(divide_factor) # 10で割る（結果は2）
print("さらに{0}で除した結果は：{1}".format(divide_factor, example3.value))
```
