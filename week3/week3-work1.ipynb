# 【課題６】図の作成

データは可視化して確認することが大切。Matplotlibやseabornを使って作図する。

- ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
- 特徴量を一つ選び、boxplotを使ってラベルごとのデータの分布を可視化する
- 特徴量を一つ選び、violinplotを使ってラベルごとのデータの分布を可視化する

boxplotとviolinplotは似た図を表示します。これらはどう違い、それぞれどのような利点があるのか説明してください。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する） 
# spicies → 0:"setosa", 1:"versicolor", 2:"virginica"
num_setosa = len(df[df.species == 0])
num_versicolor = len(df[df.species == 1])
num_virginica = len(df[df.species == 2])

data = [num_setosa, num_versicolor, num_virginica]
label = ["setosa", "versicolor", "virginica"]
textprop = {"color": "white", "weight": "bold"}

plt.pie(data, labels = label, autopct = "%1.1f%%")
```

```python
# 特徴量を一つ選び、boxplotを使ってラベルごとのデータの分布を可視化する

# 1. 特徴量にsepal_lengthを取り、ラベル毎に変数に格納する
sepal_length_setosa = df.sepal_length[df.species == 0]
sepal_length_versicolor = df.sepal_length[df.species == 1]
sepal_length_virginica = df.sepal_length[df.species == 2]

# 2. Boxplotで可視化する
sepal_length = [sepal_length_setosa, sepal_length_virginica, sepal_length_versicolor]
label = ["setosa", "versicolor", "virginica"]

plt.title("boxplot of sepal_length", color = "navy")
plt.xlabel("species", color = "navy")
plt.ylabel("sepal_length (cm)", color = "navy")
plt.boxplot(sepal_length, labels = label)
```

### Boxplotについて

Boxplotがどのような値を示しているのか、以下に簡単に記載しておく。

1. 線の上端： 最大値
2. Boxの上端： 第三四分位数
3. Boxの中心線（オレンジ色）： 第二四分位数（中央値）
4. Boxの下端：第一四分位数
5. 線の下端：最小値

```python
# 特徴量を一つ選び、violinplotを使ってラベルごとのデータの分布を可視化する

# 1. 特徴量にsepal_lengthを取り、ラベル毎に変数に格納する
sepal_length_setosa = df.sepal_length[df.species == 0]
sepal_length_versicolor = df.sepal_length[df.species == 1]
sepal_length_virginica = df.sepal_length[df.species == 2]

# 2. violinplotで可視化する
sepal_length = [sepal_length_setosa, sepal_length_virginica, sepal_length_versicolor]
label = ["setosa", "versicolor", "virginica"]

fig = plt.figure()
ax = fig.add_subplot(1,1,1) # 行数1、列数1、深さ1のsubplot（＝グラフを１つ表示）

ax.violinplot(sepal_length)
ax.set_title("violinplot of sepal_length", color = "navy")
ax.set_xlabel("species", color = "navy")
ax.set_xticks([1,2,3])
ax.set_xticklabels(label)
ax.set_ylabel("sepal_length (cm)", color = "navy")
```

### violinplotについて

violinplotについて、以下に簡単に記載しておく。

- 各データの分布密度を表現している。
- 上端：最大値
- 下端：最小値

尚、violinplotはplt.xlabelでラベル付け出来ないため、figureとaxを定義した上でaxにラベルを付ける必要がある。

### BoxplotとViolinplotの差異

- Boxplot、Violinplot共にデータの分布を可視化していることは共通している。
- Boxplotは最大値・最小値・四分位数を表示しており、各値のデータ数に関する情報は捨象されている。
- Violinplotは、データの分布密度（＝どの値のデータがどの程度存在するか）のみ表示しており、四分位数などの特徴量は示されていない。

# 【課題８】簡単な考察

- データを探索した時には、考察をすることが大切。
- 散布図行列を見ながら、3種類の花はどういう風に分けられそうかを簡単に考えて説明すること。


1. setosaの分類：

  - 可視化した散布図行列を観察すると、setosaとsetosa以外でpetal_length、petal_widthの分布に重なりがないことが確認される。
  - したがって、setosaのpetal_length、あるいはpetal_widthの最大値を境界線にすれば、setosaとsetosa以外を分類することが可能と思われる。


2. setosa以外の2種（versicolor, verginica）の分類：
  
  - 可視化した散布図行列を観察すると、versicolorとverginicaはsepal_length, sepal_widthの分布の重なりが大きいため（＝似通っているため）、sepalによる分類の精度は高くない（分布を区別する明確な境界線が引けない）と推察される。
  - 一方、versicolorとverginicaはpetal_length, petal_widthの分布の重なりは比較的小さいこと、また、petal_lengthとpetal_widthの相関係数（=0.962865）も大きい。
  - したがって、versicolorとverginicaのpetal_length、およびpetal_widthの大小を比較するにより、分類が可能と思われる。
